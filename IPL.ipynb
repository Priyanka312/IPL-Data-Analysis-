Which team has won the most matches 

```python
df.head()
```

```python
match_wins=df['match_winner'].value_counts()
```

```python
sns.barplot(y=match_wins.index,x=match_wins.values,palette='viridis')
plt.title("Most match win by team")
```

Key player Performces

1 Most "Player of the Match" Awards

```python
df.head()
```

```python
count=df['player_of_the_match'].value_counts().head(10)
```

```python
sns.barplot(y=count.index,x=count.values,palette='plasma')
plt.title("Man of The match Awards")
```

2 Top Scores

```python
df.head()
```

```python
top_scorer=df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(10)
```

```python
sns.barplot(x=top_scorer.values,y=top_scorer.index,palette='cubehelix')
plt.title("Top Scorers in IPL History")
```

Best Bowling fingures 

```python
df.head()
```

```python
df['highest_value']=df['best_bowling_figure'].apply(lambda x:x.split('--')[0]).astype(int)
```

```python
top_bowlers=df.groupby('best_bowling')['highest_value'].sum().sort_values(ascending=False).head(10).sort_values(ascending=True)
```

```python
top_bowlers.plot(kind='barh',color='teal')
```
